# Task
Perform feature engineering, preprocess the data, train a suitable model for car price prediction, evaluate its performance, and make predictions.

## Feature engineering

### Subtask:
Create new features from existing ones that could be useful for the model.


**Reasoning**:
Create new features and drop the original 'Year' column as per the instructions.



```python
df['current_year'] = 2025
df['no_year'] = df['current_year'] - df['Year']
df.drop(['Year'], axis=1, inplace=True)
display(df.head())
```

## Data preprocessing

### Subtask:
Handle categorical features and split the data into training and testing sets.


**Reasoning**:
Select and one-hot encode the categorical features, concatenate them with the original dataframe, and split the data into training and testing sets.



```python
categorical_cols = ['Fuel_Type', 'Selling_type', 'Transmission']
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
X = df_encoded.drop('Selling_Price', axis=1)
y = df_encoded['Selling_Price']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train.head())
print(y_train.head())
```

## Model selection

### Subtask:
Choose a suitable model for car price prediction.


## Model training

### Subtask:
Train the selected model on the training data.


**Reasoning**:
Train the selected RandomForestRegressor model on the training data.



```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)
```

**Reasoning**:
The error indicates that there is a string column ('Car_Name') in the training data that the model cannot process. I need to drop this column from the training and testing sets before training the model.



```python
X_train = X_train.drop('Car_Name', axis=1)
X_test = X_test.drop('Car_Name', axis=1)

model = RandomForestRegressor(random_state=0)
model.fit(X_train, y_train)
```

## Model evaluation

### Subtask:
Evaluate the model's performance on the testing data.


**Reasoning**:
Evaluate the trained model's performance on the test set by calculating and printing key regression metrics.



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
print(f'R-squared (R2): {r2:.2f}')
```

## Prediction

### Subtask:
Use the trained model to make predictions on new data.


**Reasoning**:
Use the trained model to make predictions on the test set and store them in a variable.



```python
predictions = model.predict(X_test)
print("First few predictions:")
print(predictions[:5])
```
